# 9. Wat is the probability density of price between the range 40000 to 80000?


```python
sns.distplot(dataset["Price"],kde={"color":"red"},color="green")
plt.axvline(x=dataset["Price"].mean(),color="red")
plt.axvline(x=40000,color="blue",label="low range")
plt.axvline(x=80000,color="blue")
```

```python
# here density of price around 40000 is higher than 80000
```

```python
# now to find the density of Prices between 40000 and 80000:
```

```python
mean=dataset["Price"].mean()
std=dataset["Price"].std()
from scipy.stats import norm
dist=norm(mean,std)
values=[value for value in range (40000,80000)]
[dist.pdf(value) for value in values]
# this gives the density of each price and tells us which one is more common
```

```python
pd=[dist.pdf(value) for value in values]
prob=sum(pd)
prob # this means 47 percent of the prices r between 40000 and 80000
```

# 11. wat is the pdf of inches btw 12 and 14?

```python
sns.distplot(dataset["Inches"],kde={"color":"red"},color="red")
plt.axvline(x=12,color="blue")
plt.axvline(x=14,color="blue")
```

```python
mean1=dataset["Inches"].mean()
std1=dataset["Inches"].std()
from scipy.stats import norm
dist=norm(mean1,std1)
values=[value for value in range(12,14)]
pd=[dist.pdf(value)for value in values]# gives  the density of each inch btw 12 and 14 
sum(pd)
```
